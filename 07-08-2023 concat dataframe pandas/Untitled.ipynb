```python
import pandas as pd

indian_weather = pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
indian_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
new_df = pd.concat([indian_weather,us_weather],ignore_index=True)
new_df
```

```python
new_df = pd.concat([indian_weather,us_weather],keys=['indian','us'])
new_df
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
},index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
},index=[1,0])
windspeed_df
```

```python
new_df = pd.concat([temperature_df,windspeed_df],axis=1)
new_df
```

```python
s = pd.Series(['dry','sunny','rain'], name='event')
s
```

```python
add_col = pd.concat([temperature_df,s], axis=1)
add_col
```

```python
new_df = pd.concat([add_col,windspeed_df],axis=1)
new_df
```

```python
new_df = new_df.interpolate(method='pad')
new_df
```
