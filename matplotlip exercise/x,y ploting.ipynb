```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperatute")
plt.title("Weather")
plt.plot(x,y, color="red", marker='<', linestyle="dotted", linewidth=5,label="Weather",markersize=20, alpha=0.5)
plt.grid()
plt.legend()
```
