```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df = df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Assuming X_train is a DataFrame with the two features you want to encode.
categories = [['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']]

oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)
X_train_encoded = oe.transform(X_train)
X_test_encoded = oe.transform(X_test)
```

```python
X_train_encoded
```

```python
from sklearn.preprocessing import LabelEncoder
la = LabelEncoder()
la.fit(y_train)
y_train_encoded = la.transform(y_train)
y_test_encoded = la.transform(y_test)
```
